# Scraping Produk Tokopedia (dgn halaman tertentu)

```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
keyword = 'handphone'
url =f"https://www.tokopedia.com/search?st=product&q={keyword}&srp_component_id=01.07.00.00&srp_page_id=&srp_page_title=&navsource="

driver = webdriver.Chrome()
driver.get(url)

data=[]
url_list = []
for i in range(10): #banyak page
    WebDriverWait(driver,5).until(EC.presence_of_element_located((By.CSS_SELECTOR, "#zeus-root")))
    time.sleep(2)

    for j in range(20):
        driver.execute_script("window.scrollBy(0,250)")
        time.sleep(1)

    driver.execute_script("window.scrollBy(50,0)")
    time.sleep(1)

    soup = BeautifulSoup(driver.page_source, "html.parser")
    for item in soup.findAll('div', class_='css-974ipl'):
        nama_produk = item.find('div', class_='css-3um8ox').text
        harga = item.find('div', class_='css-1ksb19c').text
        
        rtg = item.findAll('span', class_='css-t70v7i')
        if len(rtg)>0:
            rating = item.find('span', class_='css-t70v7i').text
        else:
            rating = ''

        tjl = item.findAll('span', class_='css-1duhs3e')
        if len(rtg)>0:
            terjual = item.find('span', class_='css-1duhs3e').text
        else:
            terjual = ''   
            
        url_product = item.find('a', class_='css-gwkf0u')
        product_url = url_product['href']
        
        for item2 in item.findAll('div', class_='css-1rn0irl'):
            lokasi = item2.findAll('span', class_='css-1kdc32b')[0].text
            nama_toko = item2.findAll('span', class_='css-1kdc32b')[1].text     
            
            data.append((nama_toko, lokasi, nama_produk, harga, terjual, rating, product_url))
            
    time.sleep(2)
    driver.find_element(By.CSS_SELECTOR, "button[aria-label^='Laman berikutnya']").click()
    time.sleep(3)
    
df = pd.DataFrame(data, columns=['Toko','Lokasi','Nama Produk','Harga','Terjual','rating', 'product_url'])
df.to_excel("Tokopedia scraping.xlsx", index = False)
print("Data is saved in excel")
print(df)


driver.close()
```

# Scraping Review Toko Toped (dgn halaman tertentu)

```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
url = 'https://www.tokopedia.com/brotherholicstor/review'
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
driver = webdriver.Chrome(options=options)
driver.get(url)

data = []
for i in range(0, 5):
    WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CSS_SELECTOR, "#zeus-root")))
    time.sleep(2)

    soup = BeautifulSoup(driver.page_source, "html.parser")
    containers = soup.findAll('article', attrs={'class': 'css-ccpe8t'})
    for container in containers:
        nama_produk_element = container.find('p', attrs={'data-unify': 'Typography', 'class': 'css-ra461b-unf-heading e1qvo2ff8'})
        nama_produk = nama_produk_element.text if nama_produk_element else None
        
        nama_akun_element = container.find('span', attrs={'class': 'name'})
        nama_akun = nama_akun_element.text if nama_akun_element else None

        rating_element = container.find('div', {'data-testid': 'icnStarRating'})
        rating_label = rating_element.get('aria-label')

        review_element = container.find('span', attrs={'data-testid': 'lblItemUlasan'})
        review = review_element.text if review_element else None

        data.append((nama_produk, nama_akun, review, rating_label))

    time.sleep(2)
    driver.find_element(By.CSS_SELECTOR, "button[aria-label^='Laman berikutnya']").click()
    time.sleep(3)

df = pd.DataFrame(data, columns=['nama_produk', 'nama_akun', 'review', 'rating'])
print(df)

driver.close()
```

# Scraping Produk Lazada (dgn halaman tertentu)

```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
class ScrapeLazada():
    def scrape(self):
        #pass
        url = 'https://www.lazada.co.id/tag/laptop/?q=laptop&_keyori=ss&from=input&spm=a2o4j.home.search.go.579953e0LrANDn&catalog_redirect_tag=true'
        driver = webdriver.Chrome()
        driver.get(url)
        
        data = []
        for i in range(10):
            WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CSS_SELECTOR, "#root")))
            time.sleep(2)

            soup = BeautifulSoup(driver.page_source, "html.parser")
            for item in soup.findAll('div', class_="Bm3ON"):
                product_name = item.find('div', class_="RfADt").text
                price = int(item.find('span', class_="ooOxS").text.replace("Rp", "").replace(".",""))
                location = item.find('span', class_="oa6ri").text
                
                #link = item.find('a', class_='_95X4G')
                #product_url = link.get('href')
                product_url = item.find('div', class_="RfADt").a['href']
                data.append((product_name, price, location, product_url))

            time.sleep(2)
            driver.find_element(By.CSS_SELECTOR, ".ant-pagination-next > button").click()
            time.sleep(3)   

        df = pd.DataFrame(data, columns=["product_name","price","location", "product-url"])
        df.to_excel("lazada scraping.xlsx", index = False)
        print("Data is saved in excel")
        return df
        
        driver.close()
        
sl = ScrapeLazada()
sl.scrape()
```

# Scraping produk dari Bukalapak

```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

import requests
import argparse
import random
import numpy as np
```

```python
cari = "sepatu"
products = []

for i in range(1,4): #3 halaman
    url = "https://www.bukalapak.com/products?page={}&search%5Bkeywords%5D={}".format(i,cari)    
    driver = webdriver.Chrome()
    driver.get(url)

    WebDriverWait(driver,5).until(EC.presence_of_element_located((By.CSS_SELECTOR, ".dora")))
    time.sleep(2)

    for j in range(16):
        driver.execute_script("window.scrollBy(0,250)")
        time.sleep(1)

    soup = BeautifulSoup(driver.page_source, "html.parser")

    for item in soup.findAll('div', class_='bl-product-card-new__description'):
        toko = item.find('div', class_="bl-product-card-new__store-name-and-location").find('a', class_="bl-link").text.strip() 
        lokasi = item.find('p', class_="bl-product-card-new__store-location").text.strip()
        nama = item.find('p', class_="bl-text--ellipsis__2").text.strip()
        harga = item.find('p', class_="bl-product-card-new__price").text.strip()

        rtg = item.findAll('div', class_="bl-product-card-new__star")
        if len(rtg)>0:
            rating = item.find('div', class_="bl-product-card-new__star").find('a', class_="bl-link").text.strip()
        else:
            rating ='0'

        tjl = item.findAll('p', class_="bl-product-card-new__sold-count")
        if len(tjl) >0:
            terjual = item.find('p', class_="bl-product-card-new__sold-count").text.strip()
        else:
            terjual = '0'

        products.append((toko, lokasi, nama, harga, rating, terjual))

df = pd.DataFrame(products, columns=['toko','lokasi','nama','harga','rating','terjual'])
print(df)
driver.close()
```

```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

import requests
import argparse
import random
import numpy as np
import string
```

```python
enter_input = input("Search: ")
u_i = string.capwords(enter_input)
lists = u_i.split()
word = "_".join(lists)

url = "https://en.wikipedia.org/wiki/"+word

def wikibot(url):
    url_open = requests.get(url)
    soup = BeautifulSoup(url_open.content, "html.parser")
    details = soup('table',{'class':'infobox'})
    
    for i in details:
        h = i.find_all('tr')
        for j in h:
            heading = j.find_all('th')
            detail = j.find_all('td')
            
            if heading is not None and detail is not None:
                for x,y in zip(heading,detail):
                    print("{}  :: {}".format(x.text,y.text))
                    print("------------------------------------------------")
    
    for i in range(1,10): #banyak paragraf
        print(soup('p')[i].text)
        
wikibot(url)
```
