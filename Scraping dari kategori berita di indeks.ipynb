```python
from bs4 import BeautifulSoup
import requests

"""LINK PER KATEGORI"""
link1 = [] 
kategori1 = []
url = 'https://news.detik.com/indeks' 
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

nav_item = soup.find_all(class_="grid-row content__bg mgt-16 mgb-16") #nama_class

for item in nav_item:
    nav_item1 = item.find_all(class_="nav__item") #nama_class
    for nav in nav_item1:
        link = nav.find('a')['href']
        kategori = nav.text.strip()
        kategori1.append(kategori)
        link1.append(link)
        
array_gabungan = list(zip(kategori1, link1))
#print(link1)
#print(kategori1)
print(array_gabungan)
```

```python
"""Buat untuk 10 page pertama tiap kategori"""
kategori2 = []
link2 = []
for category, link in array_gabungan:
    for i in range(1,2):
        url_full_indexs = link + '/' + str(i)
        #print(category)
        #print(url_full_indexs)
        kategori2.append(category)
        link2.append(url_full_indexs)
array_gabungan2 = list(zip(kategori2, link2))
print(array_gabungan2)
```

```python
link_berita = []
kategori3 = []
filtered_links2=[]
for category, url in array_gabungan2:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    link_artikel = soup.findAll(class_="list-content__item")
    
    for link_artikels in link_artikel:
        # Mencari elemen dengan tag 'a' dan class 'media__link'
        links = link_artikels.find('a', class_='media__link')  
        #print(links['href'])
        kategori3.append(category)
        link_berita.append(links['href'])
        
array_gabungan3 = list(zip(kategori3, link_berita))
#print(array_gabungan3)   

filtered_data = [(category, link) for category, link in array_gabungan3 if category.lower() in link.lower()]
print(filtered_data)
```

```python
"""Dari tiap berita, diambil judul, penulis, tanggal dan waktu, serta isi"""
judul = []
penulis = []
tanggal_waktu = []
isi = []
link_tiap_berita = []
kategori_berita = []

for category, j in filtered_data:
    link_tiap_berita.append(j)
    kategori_berita.append(category)

    url = j
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    print(url)
    
    #judul
    judul_artikel = soup.find(class_="detail__title").get_text(strip=True) #nama_class
    judul.append(judul_artikel)

    #penulis
    author = soup.find('div', class_='detail__author').text.split('-')[0].strip()
    penulis.append(author)

    #tanggal waktu
    date = soup.find('div', class_='detail__date').text.strip()
    tanggal_waktu.append(date)
        
    #isi
    paragraphs = soup.select('.detail__body-text p')
    combined_text = ' '.join([p.get_text(strip=True) for p in paragraphs])
    isi.append(combined_text)
```

```python
"""MEMBUAT DATAFRAME"""
import pandas as pd
df = pd.DataFrame({
        'Kategori': kategori_berita,
        'Link': link_tiap_berita,
        'Judul': judul,
        'Penulis': penulis,
        'Tanggal dan Waktu': tanggal_waktu,
        'isi': isi
    })

df      
```

```python
def link_sub_kategori1():
    hasil_kategori=[]
    url = "https://www.kompas.com/"
    req = requests.get(url)
    soup = BeautifulSoup(req.text, "html.parser")

    hasil_kategori = []
    nav_item = soup.find_all(class_="nav__item") #nama_class

    for item in nav_item:
        link_elem = item.find('a')
        if link_elem:
            link = link_elem.get('href')
            hasil_kategori.append(link)
            
    return hasil_kategori

def kategori_dan_link_berita1():
    linknya=[]
    for i in link_sub_kategori1():
        url = i
        response = requests.get(url)
        # Create a BeautifulSoup object
        soup = BeautifulSoup(response.text, "html.parser")

        # Find and print all links on the page
        link_artikel = soup.find_all(class_="row article__wrap__grid--flex col-offset-fluid") #nama_class
        
        for links in link_artikel:
            link_artikels = links.find_all('a', class_='article__link')
            for link in link_artikels:
                href = link.get('href')
                #print(href)
                linknya.append(href)

    filtered_links = [link1 for link1 in linknya if not link1.startswith('https://video.kompas.com/watch')]
    
    return filtered_links

def detail():
    judul = []
    penulis = []
    tanggal_dan_waktu = []
    isi = []
    for i in kategori_dan_link_berita1():
        url = i

        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")
        #print(url)
        
        #judul
        judul_artikel = soup.find(class_="read__title").get_text(strip=True) #nama_class
        judul.append(judul_artikel)

        #penulis
        penulis_elem = soup.find('div', id='penulis')
        if penulis_elem is not None:
            nama_penulis = penulis_elem.find('a').text
        else:
            nama_penulis = "Unknown"
        penulis.append(nama_penulis)
        
        #tanggal waktu
        time_elem = soup.find('div', class_='read__time')
        teks_lengkap = time_elem.text
        tanggal_waktu = teks_lengkap.split('-')[1].strip().replace("Diperbarui ","")
        tanggal_dan_waktu.append(tanggal_waktu)
        #print(tanggal_waktu)
        
        #isi
        isi_artikel = soup.find(class_="read__content").get_text(strip=True) #nama_class
        isi.append(isi_artikel)
                
    df = pd.DataFrame({
        'Link berita': i,
        'Judul': judul,
        'Penulis': penulis,
        'Tanggal dan Waktu': tanggal_waktu,
        'isi': isi
    })
    #return judul, penulis, tanggal_waktu, isi
    return df
```

```python
cari = "sepatu"
products = []

for i in range(1,4): #3 halaman
    url = "https://www.bukalapak.com/products?page={}&search%5Bkeywords%5D={}".format(i,cari)    
    driver = webdriver.Chrome()
    driver.get(url)

    WebDriverWait(driver,5).until(EC.presence_of_element_located((By.CSS_SELECTOR, ".dora")))
    time.sleep(2)

    for j in range(16):
        driver.execute_script("window.scrollBy(0,250)")
        time.sleep(1)

    soup = BeautifulSoup(driver.page_source, "html.parser")

    for item in soup.findAll('div', class_='bl-product-card-new__description'):
        toko = item.find('div', class_="bl-product-card-new__store-name-and-location").find('a', class_="bl-link").text.strip() 
        lokasi = item.find('p', class_="bl-product-card-new__store-location").text.strip()
        nama = item.find('p', class_="bl-text--ellipsis__2").text.strip()
        harga = item.find('p', class_="bl-product-card-new__price").text.strip()

        rtg = item.findAll('div', class_="bl-product-card-new__star")
        if len(rtg)>0:
            rating = item.find('div', class_="bl-product-card-new__star").find('a', class_="bl-link").text.strip()
        else:
            rating ='0'

        tjl = item.findAll('p', class_="bl-product-card-new__sold-count")
        if len(tjl) >0:
            terjual = item.find('p', class_="bl-product-card-new__sold-count").text.strip()
        else:
            terjual = '0'

        products.append((toko, lokasi, nama, harga, rating, terjual))

df = pd.DataFrame(products, columns=['toko','lokasi','nama','harga','rating','terjual'])
print(df)
driver.close()
```
